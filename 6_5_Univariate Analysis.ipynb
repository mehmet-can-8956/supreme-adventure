```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.figsize'] = (8,5.5)
```

```python
us_edu = pd.read_csv("datasets_95317_1078789_states_all.csv")
```

```python
us_edu.drop(us_edu[us_edu["STATE"]=="NATIONAL"].index, axis=0, inplace=True)
```

```python
us_edu.drop(us_edu[us_edu["STATE"]=="DODEA"].index, axis=0, inplace=True)
```

```python
states=us_edu["STATE"].unique()
```

```python
for eachcolumn in ("ENROLL","TOTAL_REVENUE","FEDERAL_REVENUE","STATE_REVENUE","LOCAL_REVENUE","TOTAL_EXPENDITURE",
                   "INSTRUCTION_EXPENDITURE","SUPPORT_SERVICES_EXPENDITURE","OTHER_EXPENDITURE","CAPITAL_OUTLAY_EXPENDITURE",
                   "GRADES_PK_G","GRADES_KG_G","GRADES_4_G","GRADES_8_G","GRADES_12_G","GRADES_1_8_G","GRADES_9_12_G",
                   "GRADES_ALL_G","AVG_MATH_4_SCORE","AVG_MATH_8_SCORE","AVG_READING_4_SCORE","AVG_READING_8_SCORE"):
    for state in states:
        us_edu[us_edu["STATE"]==state][eachcolumn].fillna(us_edu[us_edu["STATE"]==state][eachcolumn].mean())
```

```python
us_edu = pd.read_csv("datasets_95317_1078789_states_all.csv")
```

```python
us_edu.drop(us_edu[us_edu["STATE"]!="CALIFORNIA"].index, axis=0, inplace=True)
```

```python
cali_total=us_edu[["STATE","YEAR","TOTAL_REVENUE","TOTAL_EXPENDITURE"]].sort_values(by=['YEAR'])
```

```python
cali_total["TOTAL_REVENUE"].interpolate(method ='linear', limit_direction ='both', inplace=True)
cali_total["TOTAL_EXPENDITURE"].interpolate(method ='linear', limit_direction ='both', inplace=True)
```

```python
title_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
axis_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }
pd.DataFrame(cali_total).plot(x = 'YEAR', y=["TOTAL_REVENUE","TOTAL_EXPENDITURE"])
plt.title('Total Revenue & Expenditure through Years', fontdict=title_font )
plt.ylabel('Total Revenue & Expenditure', fontdict=axis_font )
plt.xlabel('Years', fontdict=axis_font )
plt.legend(('Revenue', 'Expenditure'))
plt.show()
```
