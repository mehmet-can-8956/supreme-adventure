```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv("train.csv", index_col=0)
titanic
```

```python
survived_30=titanic[(titanic['Survived']==1) & (titanic['Age']>30)]
survived_30
```

```python
print("the survival rate for all passengers under 30 years old:",survived_30.shape[0]/titanic.shape[0])
```

```python
survived_male=titanic[(titanic['Survived']==1) & (titanic['Sex']=='male')]
survived_male
```

```python
survived_female=titanic[(titanic['Survived']==1) & (titanic['Sex']=='female')]
survived_female
```

```python
print("the survival rate for males  :",survived_male.shape[0]/(survived_male.shape[0]+survived_female.shape[0]))
print("the survival rate for females:",survived_female.shape[0]/(survived_male.shape[0]+survived_female.shape[0]))
```
