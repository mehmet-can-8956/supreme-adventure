```python
google = pd.read_csv("GOOGL.csv", index_col=0)
google
```

```python
google['Adj Close'].plot(figsize=(10,4))
plt.xlabel("Date")
plt.ylabel("Adjusted Price")
plt.title("Google Price data")
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
google['Adj Close'].plot.hist(bins = 60)
ax1.set_xlabel("Returns %")
ax1.set_ylabel("Percent")
ax1.set_title("Google returns data")
plt.show()
```

```python
google_daily_returns = google['Adj Close'].pct_change()
print(google_daily_returns.head())
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
google_daily_returns.plot.hist(bins = 60)
ax1.set_xlabel("Daily returns %")
ax1.set_ylabel("Percent")
ax1.set_title("Google daily returns data")
plt.show()
```
