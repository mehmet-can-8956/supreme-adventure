```python
import numpy as np
```

```python
sample=np.load("earthquakes_sample.npy")
sample
```

```python
sample1 = np.arange(100,dtype=float).reshape(20,5)
for i in range(len(sample)):
    for j in range(len(sample[0,:])):
         sample1[i,j]=float(sample[i,j])
sample1
```

```python
for i in range(len(sample1[0,:])):
    print("Standart deviation of column",i,":", np.std(sample1[:,i]))
    print("Mean of column              ",i,":", np.mean(sample1[:,i]))
```
