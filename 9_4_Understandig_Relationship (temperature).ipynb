```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
import statsmodels.formula.api as smf

# Display preferences.
%matplotlib inline
pd.options.display.float_format = '{:.3f}'.format

import warnings
warnings.filterwarnings(action="ignore")

weather = pd.read_csv("weatherHistory.csv")

weather.head(1)
```

```python
weather.rename(columns={"Temperature (C)": "temperature","Wind Speed (km/h)":"windspeed","Apparent Temperature (C)":"apparenttemp","Humidity":"humidity"}, inplace=True)
```

```python
weather=weather[["temperature","humidity","windspeed","apparenttemp"]]
weather.head(3)
```

```python
weather["temp_dif"]=weather["temperature"]-weather["apparenttemp"]
weather.head(3)
```

```python
weather.corr()
```

```python
# Y is the target variable
Y = weather["temp_dif"]

# X is the feature set
X = weather[["humidity","windspeed"]]

# We add constant to the model as it's a best practice
# to do so everytime!
X = sm.add_constant(X)

# We fit an OLS model using statsmodels
results = sm.OLS(Y, X).fit()

# We print the summary results
display(results.summary())
```
