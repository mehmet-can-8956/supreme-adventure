```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
pop1 = np.random.normal(10, 0.2, 1000)
pop2 = np.random.normal(10, 0.2, 1000)
```

```python
sample1_means = []
sample1_stds  = []
for i in range(1,1000):
    sample1 = np.random.choice(pop1, 100, replace=True)
    sample1_means.append(np.mean(sample1))
    sample1_stds.append(np.std(sample1))
```

```python
sample2_means = []
sample2_stds  = []
for i in range(1,50):
    sample2 = np.random.choice(pop2, 100, replace=True)
    sample2_means.append(np.mean(sample2))
    sample2_stds.append(np.std(sample2))
```

```python
plt.figure(figsize=(8,5))

plt.hist(sample1_means, alpha=0.5, label='Sample 1 Means for 1000 times', bins=50) 
plt.hist(sample2_means, alpha=0.5, label='Sample 2 Means for 50 times', bins=50) 
plt.legend(loc='upper right') 
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(sample1_stds, alpha=0.5, label='Sample 1 Stds for 1000 times', bins=50) 
plt.hist(sample2_stds, alpha=0.5, label='Sample 2 Stds for 50 times', bins=50) 
plt.legend(loc='upper right') 
plt.show()
```

```python
# Interpretig the graphics
# when we analyse the outputs, calculating 1000 times average
# and deviation made the samples closer. It is obvious that
# the Central Limit Theorem seems to have worked.
```
